# Clustering with K-means: an application to Image Compression

In this Notebook we will use K-means to find $K$ representative points in a $n\times m \times 3$ color image. If each color is coded with 8 bits, then the image requires $n\times m \times 3$ bytes of memory.

We will compress the image so that the color of each pixel in the image (a 3-dimensional vector) will be given by the color of the closest centroid. In this way, the compressed image requires $n\times m \times \frac{\log_2(K)}{8}$ bytes of memory, plus the memory to store the color of each of the $K$ centroids ($K \times 3$ bytes). This represents a compression rate of  
$$ \eta=\frac{n ~m ~ \frac{\log_2(K)}{8} + 3K}{3~m ~ n}=\frac{\log_2(K)}{21}+\frac{K}{m~n}$$

Along with the compression rate, we will analyze the quality of the reconstruction image (e.g. distortion) by computing the average squared distance between pixels in the original image and the reconstructed one.

We transform the image from unit8 (0-255) to (0-1) values

```python
val_min=np.min(image_double)
val_max=np.max(image_double)

image_double=(image_double - val_min) / (val_max - val_min)
```
